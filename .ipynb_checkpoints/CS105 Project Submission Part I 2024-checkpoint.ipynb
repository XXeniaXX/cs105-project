### 2. Data pre-processing [35% of Part I]

**a.** For each variable, determine the percentage of missing data. For any column with missing data, describe how you resolve the issue. Clearly state any assumption you made.

```python
import pandas as pd
data = pd.read_csv(r"housing.csv")
df = pd.DataFrame(data, columns=["ID", "Crime_Rate", "Residential_Area", "Non_Retail", "River", "Pollutants", "Rooms", "Old_Housing", "Distance_Office", "Highway_Access", "Tax_Rate", "Student_Teacher_Ratio", "Low_Income", "Price"])
# Array of name of each column 
columns = ["ID", "Crime_Rate", "Residential_Area", "Non_Retail", 
           "River", "Pollutants", "Rooms", "Old_Housing", "Distance_Office", 
           "Highway_Access", "Tax_Rate", "Student_Teacher_Ratio", "Low_Income", "Price"]


# Table header
print("{:22s} {:25s}".format("Category", "Percentage of missing data"))


# Use Match-Case statements to iterate through each row of every column 
for category in columns:
    match category:
        case "ID":
            numOfMissing = len(df[pd.isnull(df.ID)])
        case "Crime_Rate":
            numOfMissing = len(df[pd.isnull(df.Crime_Rate)])
        case "Residential_Area":
            numOfMissing = len(df[pd.isnull(df.Residential_Area)])
        case "Non_Retail":
            numOfMissing = len(df[pd.isnull(df.Non_Retail)])
        case "River":
            numOfMissing = len(df[pd.isnull(df.River)])
        case "Pollutants":
            numOfMissing = len(df[pd.isnull(df.Pollutants)])
        case "Rooms":
            numOfMissing = len(df[pd.isnull(df.Rooms)])
        case "Old_Housing":
            numOfMissing = len(df[pd.isnull(df.Old_Housing)])
        case "Distance_Office":
            numOfMissing = len(df[pd.isnull(df.Distance_Office)])
        case "Highway_Access":
            numOfMissing = len(df[pd.isnull(df.Highway_Access)])
        case "Tax_Rate":
            numOfMissing = len(df[pd.isnull(df.Tax_Rate)])
        case "Student_Teacher_Ratio":
            numOfMissing = len(df[pd.isnull(df.Student_Teacher_Ratio)])
        case "Low_Income":
            numOfMissing = len(df[pd.isnull(df.Low_Income)])
        case "Price":
            numOfMissing = len(df[pd.isnull(df.Price)])
        case _:
            numOfMissing = 0
    
    # Calculate and display percentage of missing data for each column 
    missingPercentage = (numOfMissing/len(df))*100
    print("{:22s} {:.2f}%".format(category, missingPercentage))
    
print(df.info())
```
